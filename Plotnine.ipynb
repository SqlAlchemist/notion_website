# Using R ggplot2 Package in Python With plotnine Library

![ggplot](https://financesonline.com/uploads/2019/09/ggplot2-logo1.png "ggplot")
![plotnine](http://download.broadview.com.cn/Original/21019700a5ade45e047d "plotnine")

`ggplot2` is based on Hadley Wickham's [Layered Grammar of Graphics](http://vita.had.co.nz/papers/layered-grammar.pdf), which provides a systematic way to describe the components of a statistical graphic.

`ggplot2` is a declarative language, which implies to specify the *"what you want"* to plot, rather than *"the how"*.

`ggplot2` has a wide range of features, including:

 * Support for a variety of data types (data frames, matrices, and lists)
 * A wide variety of geometric objects (points, lines, bars, and histograms)
 * A variety of statistical transformations (aggregation, smoothing, and binning)
 * A powerful system for customizing plots

## A Layered Grammar of Graphics

There are 03 compulsory components for building visualizations with `plotnine`:

1. Data
2. Aesthetics
3. Geometric objects

## Environment settings

```python
from plotnine import *
from plotnine.data import *
import numpy as np
import pandas as pd
import polars as pl
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Datasets

### Unemployment and population

```python
economics.head()
```

```python
chart = (
    ggplot(data=economics)
    + aes(x='date', y='uempmed')
    + geom_line(color='#670000', size=1)
    + labs(title='US unemployed duration (weeks)')
    + theme(plot_title=element_text(size=17, family='Helvetica Neue', face='bold'))
    + xlab('')
    + ylab('Unemployed duration (median)')
    #+ theme_538() # set fivethirtyeight theme
)
```

```python
chart
```

```python
(
    ggplot(economics)
    + aes(x="date", y="pop")
    + geom_line(color='#670000', size=1)
    + labs(title="US Population over time", y="Population")
    + scale_x_timedelta(name="Years since 1970")
)
```

### Motors and cylinders

```python
mpg.head()
```

```python
(
    ggplot(mpg)
    + aes(x="class")
    + geom_bar(color='white', fill='#670000',)
    + ggtitle('Bar chart of car class')
)
```

```python
(
    ggplot(mpg)
    + aes(x="class")
    + geom_bar(fill='#670000',)
    + coord_flip()
    + ggtitle('Horizontal bar chart of car class')
)
```

```python
(
    ggplot(mpg)
    + aes(x="class", y="hwy")
    + geom_point(color='#670000', size=3)
    + ggtitle('Scatter chart of car class and speed')
)
```

```python
(
    ggplot(mpg)
    + aes(x="class", y="hwy")
    + geom_boxplot(color='#670000',)
    + ggtitle('Boxplot of cars and highway top speed')
)
```

```python
(
    ggplot(mpg)
    + aes(x="class", y="cty")
    + geom_violin(color='#670000',)
    + ggtitle('Violin plot of car class and cty')
)
```

## Statistical transformations

```python
huron.head()
```

```python
(
    ggplot(huron)
    + aes(x="level")
    + stat_bin(bins=8)
    + geom_bar()
    + ggtitle('Distribution plot of Huron Lake\'s water level')
)
```

```python
(
    ggplot(huron)
    + aes(x="level")
    + geom_histogram(bins=10, color='#670000', fill='#a6a6a6')
    + ggtitle('Distribution plot of Huron Lake\'s water level')
)
```

```python
(
    ggplot(huron)
    + aes(x="factor(decade)", y="level")
    + geom_boxplot(color='#670000',)
    + ggtitle('Boxplot of Huron Lake\'s water level by decade')
)
```
